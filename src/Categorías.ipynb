**Categorías**



Mediante el siguiente código se genera una representación visual de una leyenda de categorías. Cada categoría tiene un color progresivo y legible. La matriz de fondo contiene los números de categoría, y los nombres de las categorías se colocan sobre ella con colores apropiados. El resultado es una leyenda colorida y comprensible con las 12 clases que hay en las imagenes de las laparoscopias.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
category_height = 100
legend_matrix = np.zeros((13 * category_height, 800))
for i in range(13):
    legend_matrix[i * category_height: (i + 1) * category_height] = i
```

```python
category_names = {
    0: 'Background',
    1: 'Abdominal Wall',
    2: 'Liver',
    3: 'Gastrointestinal Tract',
    4: 'Fat',
    5: 'Grasper',
    6: 'Connective Tissue',
    7: 'Blood',
    8: 'Cystic Duct',
    9: 'L-hook Electrocautery',
    10: 'Gallbladder',
    11: 'Hepatic Vein',
    12: 'Liver Ligament',
}

category_colors = [
    '#FF5733', '#FF9D33', '#FFD433', '#E6FF33', '#A3FF33',
    '#5AFF33', '#33FF48', '#33FF8C', '#33FFC1', '#33E6FF',
    '#33A3FF', '#335AFF', '#3348FF'
]
```

```python
plt.figure(figsize=(8, 6))
plt.title('Categoría de la Imagen')

plt.imshow(legend_matrix, cmap='afmhot', norm=plt.Normalize(vmin=0, vmax=12))


for i in range(13):
    plt.text(400, i * category_height + category_height / 2, f"{list(category_names.values())[i]}",
             ha="center", va="center", color=category_colors[i], fontsize='medium', weight='bold')  # Cambiar el peso de la letra a negrita

plt.axis("off")
plt.show()
```
